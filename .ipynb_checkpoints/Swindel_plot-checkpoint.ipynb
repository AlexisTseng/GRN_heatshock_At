```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
import sys
import math
```

```python
dtype_dict = {'gene':'str', 'Time':'float', 'group': 'str', 'Signal': 'float', 'P-Value': 'float'}
sd_data = pd.read_csv('Swindell_HSR_2.csv', dtype = dtype_dict)
print(sd_data)
```

```python
sd_data.info()
#sd_data.describe()
#sd_data['gene'] = sd_data['gene'].astype(str)
sd_data['gene'].unique()
```

```python
HsfA2 = ['HsfA2']
HsfA1 = [ 'HsfA1a ', 'HsfA1b', 'HsfA1e','HsfA1d ']
Hsfb = ['HsfB3', 'HsfB4', 'HsfB2b', 'HsfB2a', 'HsfB1']
Hsp90 = ['Hsp90-6', 'Hsp90-4', 'Hsp90-5', 'Hsp90-7', 'Hsp90-1', 'Hsp90-2']
```

```python
grouped_data = sd_data.groupby('group')
```

```python
print(sd_data.dtypes)
```

```python
A2_control = sd_data[(sd_data['gene'] == 'HsfA2') & (sd_data['group'] == 'Control')]
print(A2_control['Time'])
```

```python

fig, ax = plt.subplots(ncols = 2, nrows = 4, figsize = (10, 15))
for i, pro in enumerate([HsfA1, HsfA2, Hsfb, Hsp90]):
    for y_col in pro:
        control_df = sd_data[(sd_data['gene'] == y_col) & (sd_data['group'] == 'Control')]
        ax[i,0].scatter(control_df['Time'], control_df['Signal'], alpha=0.8)
        ax[i,0].plot(control_df['Time'].unique(), control_df.groupby('Time')['Signal'].mean(), label=y_col, linewidth=1, alpha=0.8)
        ax[i,0].set_xlabel('Time (hour) post HS')
        ax[i,0].set_ylabel('Transcript abundance, northern blot')
        ax[i,0].set_title('Control')
        #ax[i,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
        heat_df = sd_data[(sd_data['gene'] == y_col) & (sd_data['group'] == 'Heat')]
        ax[i,1].scatter(heat_df['Time'], heat_df['Signal'], alpha=0.8)
        ax[i,1].plot(heat_df['Time'].unique(), heat_df.groupby('Time')['Signal'].mean(), label=y_col, linewidth=1, alpha=0.8)
        ax[i,1].set_xlabel('Time (hour) post HS')
        ax[i,1].set_ylabel('Transcript abundance, northern blot')
        ax[i,1].set_title('heat')
        ax[i,1].legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)


plt.tight_layout()
plt.show()

```
